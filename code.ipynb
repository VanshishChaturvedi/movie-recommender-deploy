```python
movies = movies.merge(credits, on=['movie_id', 'title'])
movies.head()
```

```python
# now we will drop some columns and keep some:
# to keep -
# genres
# movie_id
# keywords
# title
# overview
# production_companies
# release_date
# revenue
# cast
# crew
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'production_companies', 'release_date', 'revenue', 'cast', 'crew']]
```

Handling Null values

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

Making data in right format

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  result = []
  for i in ast.literal_eval(obj):
    result.append(i["name"])
  return result
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast
def convert_c(obj):
  counter = 0
  result = []
  for i in ast.literal_eval(obj):
    if counter<3:
      result.append(i["name"])
      counter += 1
    else:
      break
  return result
```

```python
movies["cast"] = movies["cast"].apply(convert_c)
```

```python
movies.head()
```

```python
import ast
def convert_cr(obj):
  result = []
  for i in ast.literal_eval(obj):
    if i['job'] in ["Director", "Writer", "Original Music Composer", "Screenplay"]:
      result.append(i["name"])
  return result
```

```python
movies["crew"] = movies["crew"].apply(convert_cr)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies["production_companies"] = movies["production_companies"].apply(convert)
```

```python
movies.head()
```

```python
def get_decade(date):
    if pd.isnull(date):
        return "Unknown"
    year = pd.to_datetime(date).year
    decade = (year // 10) * 10
    return f"{decade}s"

movies['release_date'] = movies['release_date'].apply(get_decade)
```

```python
movies.head()
```

```python
revenue_bins = [0, 100000000, 500000000, 1000000000, numpy.inf]
revenue_labels = ['Under 100M', '100M+', '500M+', '1B+']

movies['revenue'] = pd.cut(movies['revenue'], 
                                bins=revenue_bins, 
                                labels=revenue_labels, 
                                include_lowest=True)

print(movies[['revenue', 'revenue']].head())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['revenue'] = movies['revenue'].astype(str).apply(lambda x: [x])
movies['release_date'] = movies['release_date'].astype(str).apply(lambda x: [x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['production_companies'] + movies['genres'] + movies['release_date'] + movies['revenue']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = movies[movies['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
